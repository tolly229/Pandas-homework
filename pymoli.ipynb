```python
import pandas as pd
import numpy as np

file_to_load = "purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
numberOfPlayers = len(purchase_data["SN"].unique())
df1 = pd.DataFrame({"Total Players": [numberOfPlayers]})
df1
```

```python
numberOfUniqueItems = len(purchase_data["Item ID"].unique())
```

```python
numberOfItems = purchase_data["Item Name"].value_counts()
```

```python
numberOfItemsPurchases = purchase_data["Item ID"].value_counts()
```

```python
averagePrice= purchase_data["Price"].mean()
```

```python
total = purchase_data["Price"].sum()
```

```python
df2 = pd.DataFrame({"Number of Unique Items": [numberOfUniqueItems], "Average Price": [round(averagePrice,2)],"Number of Purchases":[numberOfItemsPurchases.sum()],"Total Revenue":[total]})
df2["Average Price"]= df2["Average Price"].map("${:,.2f}".format)
df2["Total Revenue"]= df2["Total Revenue"].map("${:,.2f}".format)
df2
```

```python
total_players=purchase_data[["SN","Gender","Age"]]
total_players = total_players.drop_duplicates()
```

```python
grouped_gender_df = total_players.groupby(['Gender'])
gender_count = grouped_gender_df.count()["SN"]
gender_percentage = gender_count / numberOfPlayers *100
df3 = pd.DataFrame({"Total Count":gender_count, "Percentage of Players":gender_percentage.round(2)})
df3
```

```python
grouped_gender_count= purchase_data.groupby(["Gender"])
gender_purchase_count = grouped_gender_count.count()["Purchase ID"]
gender_average_price = grouped_gender_count.mean()["Price"]
gender_total_purchase = grouped_gender_count.sum()["Price"]
gender_average_total = gender_total_purchase/gender_count 
df4 = pd.DataFrame({"Purchase Count":gender_purchase_count,"Average Purchase Price":gender_average_price.round(2),"Total Purchase Value":gender_total_purchase,"Avg Total Purchase per Person":gender_average_total.round(2)})
df4["Average Purchase Price"]= df4["Average Purchase Price"].map("${:,.2f}".format)
df4["Total Purchase Value"]= df4["Total Purchase Value"].map("${:,.2f}".format)
df4["Avg Total Purchase per Person"]= df4["Avg Total Purchase per Person"].map("${:,.2f}".format)
df4
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
total_players["Age Group"] = pd.cut(total_players["Age"], bins, labels=age)
age_count= total_players["Age Group"].value_counts()
age_percentage = age_count/numberOfPlayers *100
index = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
df5 = pd.DataFrame({"Total Count":age_count, "Percentage of Players": age_percentage.round(2)},index=index)
df5
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age)
grouped_age_count= purchase_data.groupby("Age Group")
age_purchase_count = grouped_age_count.count()["Purchase ID"]
age_average_price = grouped_age_count.mean()["Price"]
age_total_purchase = grouped_age_count.sum()["Price"]
age_average_purchase_person = age_total_purchase/age_count
index = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
df5 = pd.DataFrame({"Purchase Count":age_purchase_count,"Average Purchase Price":age_average_price.round(2),"Total Purchase Value":age_total_purchase,"Avg Total Purchase per Person":age_average_purchase_person.round(2)})
df5["Average Purchase Price"]= df5["Average Purchase Price"].map("${:,.2f}".format)
df5["Total Purchase Value"]= df5["Total Purchase Value"].map("${:,.2f}".format)
df5["Avg Total Purchase per Person"]= df5["Avg Total Purchase per Person"].map("${:,.2f}".format)
df5
```

```python
grouped_sn_count = purchase_data.groupby("SN")
sn_purchase_count = grouped_sn_count.count()["Purchase ID"]
sn_average_purchase=grouped_sn_count.mean()["Price"]
sn_total_purchase= grouped_sn_count.sum()["Price"]
df6 = pd.DataFrame({"Purchase Count":sn_purchase_count,"Average Purchase Price":sn_average_purchase.round(2),"Total Purchase Value":sn_total_purchase})
df6 = df6.sort_values(["Total Purchase Value"], ascending=False)
df6["Average Purchase Price"]= df6["Average Purchase Price"].map("${:,.2f}".format)
df6["Total Purchase Value"]= df6["Total Purchase Value"].map("${:,.2f}".format)
df6.head()
```

```python
popular_items = purchase_data.loc[:,["Item ID","Item Name","Price"]]
grouped_popular_items = popular_items.groupby(["Item ID","Item Name"])
popular_items_count = grouped_popular_items.count()["Price"]
popular_items_average = grouped_popular_items.mean()["Price"]
popular_items_total = grouped_popular_items.sum()["Price"]
df7 = pd.DataFrame({"Purchase Count":popular_items_count,"Item Price":popular_items_average.round(2),"Total Purchase Value":popular_items_total})
df7 = df7.sort_values(["Purchase Count"], ascending=False)
df7["Item Price"]= df7["Item Price"].map("${:,.2f}".format)
df7.head()
```

```python
df8 = df7.sort_values(["Total Purchase Value"], ascending=False)
df8["Total Purchase Value"]= df8["Total Purchase Value"].map("${:,.2f}".format)
df8.head()
```
